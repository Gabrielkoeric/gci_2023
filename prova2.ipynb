```python
import pandas as pd

base = pd.read_excel('BRCidades.xlsx')

base.head()
```

```python
am = base.loc[base['estado'] == 'PR']
am
```

```python
data = am[
    [
        'area_territorial',
        'populacao_estimada',
        'densidade_demografica',
        'escolarizacao',
        'carros',
        'cidade',
        'latitude',
        'longitude',
        'unidades_rurais',
        'area_plantada',
        'producao_graos'
    ]
]
```

```python
data.query("carros > 100000")
```

```python
data.query("cidade == 'Curitiba'")
```

```python
import seaborn as sns

sns.regplot(data=data, x='area_plantada', y='producao_graos')
```

```python
sns.boxplot(data=data, x='densidade_demografica', y='carros')
```
